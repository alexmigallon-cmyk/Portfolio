### **Variables categóricas.**

En caso de tener más información acerca de la causa de la falta de los valores podríamos imputarlos de manera más adecuada, ya sea asignando la categoría correcta o creando una nueva categoría específica que refleje esa ausencia."

### Variables numéricas
Para este tipo de variables, haremos primero un pequeño estudio para saber que estadística usar y si usar la global o por grupos.

```python
df.groupby('city_name')['elevation_m'].apply(lambda x: x.isna().sum())
```

```python
df.groupby('city_name')['elevation_m'].describe()
```

```python
cities_with_nulls = df.loc[df['elevation_m'].isna(), 'city_name'].unique()
elev_stats = (
    df[df['city_name'].isin(cities_with_nulls)]
    .groupby('city_name')['elevation_m']
    .agg(['count','min','median','max', 'var', 'mean'])
    .sort_values('median')
)
print(elev_stats)
```

```python
medianas = (df[df['elevation_m'].notna()]
            .groupby('city_name')['elevation_m'].median())
cv = medianas.std() / medianas.mean()

print("CV medianas por ciudad:", round(cv, 3))
if cv > 0.15:
    print("→ Diferencias relevantes entre ciudades: imputa por ciudad (mediana por ciudad).")
else:
    print("→ Variabilidad baja: puedes imputar con la mediana global.")
```

```python
df['elevation_m'] = df.groupby('city_name')['elevation_m']\
                      .transform(lambda s: s.fillna(s.median()))
```

```python
df.isna().sum()
```

```python
def diagnose_strategy(df, col, group='city_name'):
    s = df[df[col].notna()].groupby(group)[col]
    stats = pd.DataFrame({
        'count':  s.size(),
        'mean':   s.mean(),
        'median': s.median(),
        'p10':    s.quantile(0.10),
        'p90':    s.quantile(0.90),
        'trimmed_mean_10': s.apply(lambda x: x[(x>=x.quantile(0.10)) & (x<=x.quantile(0.90))].mean())
    })
    stats['abs_diff_mean_median'] = (stats['mean'] - stats['median']).abs()
    stats['rel_diff_mean_median'] = stats['abs_diff_mean_median'] / (stats['median'].abs() + 1e-9)

    cv_medians = stats['median'].std() / (stats['median'].mean() + 1e-9)
    share_big_diff = (stats['rel_diff_mean_median'] > 0.10).mean()

    return stats.sort_values('abs_diff_mean_median', ascending=False), cv_medians, share_big_diff
```

```python
num_targets = ['drainage_density_km_per_km2', 'storm_drain_proximity_m']

for col in num_targets:
    df[col] = pd.to_numeric(df[col], errors='coerce')

for col in num_targets:
    stats_city, cv_city, share_city = diagnose_strategy(df, col, group='city_name')
    print(f"\n=== {col} | Diagnóstico por ciudad ===")
    print(f"CV medianas entre ciudades: {cv_city:.3f}")
    print(f"% ciudades con |media-mediana| > 10%: {share_city:.1%}")
    print("Sugerencia:",
          "MEDIANA por CIUDAD" if (share_city > 0.30 or cv_city > 0.15) else "Mediana GLOBAL")
```

En el primer caso, vemos que no hay casi diferencia entre media y mediana ya que ninguna ciudad hay más de un 10% de diferencia, además vemos que el CV entre medianas es muy bajo, podríamos optar tanto por la media como por la mediana global para imputar, en este caso decidimos usar la mediana, pero con la media las diferencias serían mínimas. Por otro lado para el segundo caso vemos que la diferencia entre media y mediana ya es mucho más alta y además y además el CV entre medianas es también más alto, por lo tanto, en este caso si que parece que la solución más óptima es escoger la mediana por grupos para imputar.

```python
df['drainage_density_km_per_km2'] = df['drainage_density_km_per_km2']\
    .fillna(df['drainage_density_km_per_km2'].median())
```

```python
df['storm_drain_proximity_m'] = df.groupby('city_name')['storm_drain_proximity_m']\
    .transform(lambda s: s.fillna(s.median()))
```
